```

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
import nltk
nltk.download('wordnet')
import re
nltk.download('stopwords')
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import embeddings
nltk.download('punkt')
from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from keras.preprocessing import sequence
import numpy as np
from keras.models import  Sequential
from keras.layers import Embedding,Dropout,Dense,LSTM
import gensim
```

```
df.head()
```

```
df = df.dropna(axis='rows')
df.reset_index(inplace = True)
```

```
corpus = []
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['Reviews'][i])
    review = review.lower()
    review = review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in (stopwords.words("english"),"samsung","blu","apple","lg","blackberry","nokia","motorola","htc","cnpgd","otterbox","sony","posh mobile","huawei","lg electronics","polaroid","zte","alcatel","asus","verykool","casio","lenovo","ultrabox","rca","figo","pantech","samsung korea ltd","amazon.com")]
    review = ' '.join(review)
    corpus.append(review)
df["new_review"] = corpus
df.head()
```

```
# from sklearn.feature_extraction.text import TfidfVectorizer
# cv = TfidfVectorizer()
# onehot_repr = cv.fit_transform(corpus).toarray()
```

```
voc_size = 10000
from tensorflow.keras.preprocessing.text import one_hot
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```
sent_length=140
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```
df['labels']=np.where(df['Rating'] > 3, 1, 0)
df.head()
```

```
y=df['labels']
```

```
X = df["new_review"]
```

```
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```
X_final.shape,y_final.shape
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```
embedding_vector_features=100
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=12,batch_size=32)
```

```
y_pred=model.predict_classes(X_test)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
